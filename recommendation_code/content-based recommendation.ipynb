# Content-Based recommendation system

### <span style="color:red">userInput은 user에 따라 변경 가능</span>

```python
userInput = [{'course' : '인천대공원,현대프리미엄아울렛송도점,오크우드프리미어인천오크레스토랑', 'rating' : 9},
            {'course' : '담양천,담빛예술창고,담양국수거리', 'rating' : 6},
            {'course' : '부산정란각,초량이바구길,부평깡통시장,보수동책방골목,국제시장,자갈치시장,흰여울문화마을,감천문화마을', 'rating' : 5},
            {'course' : '울산대왕암공원,젊음의거리,짚신매우갈비찜,장생포고래박물관', 'rating' : 4},
            {'course' : '설악산더퀸,설악산국립공원케이블카,김영래할머니순두부본점,테디베어팜,바우지움조각미술관', 'rating' : 5}]
```

```python
user_df = pd.DataFrame(userInput)
user_df
```

```python
inputId = course_df[course_df['course'].isin(user_df['course'].tolist())]
result = pd.merge(inputId, user_df)
result
```

```python
result_df = result[['courseid', 'course', 'rating']]
result_df
```

```python
result_detail = result[['특색', '물', '나무', '유적지', '맛집여행', '호캉스', '감성카페', '애견동반ㅇ', '봄', '여름', '가을', '겨울']]
result_detail
```

```python
result_df['rating']
```

```python
userProfile = result_detail.transpose().dot(result_df['rating'])
userProfile
```

```python
courseTable = course_df.set_index(course_df['courseid'])
courseTable = courseTable[['특색', '물', '나무', '유적지', '맛집여행', '호캉스', '감성카페', '애견동반ㅇ', '봄', '여름', '가을', '겨울']]
courseTable.head()
```

```python
courseTable.shape
```

### <span style="color:red">course특징에 대해서 가중평균 취하기</span>

```python
recoTable_df = ((courseTable * userProfile).sum(axis = 1)) / (userProfile.sum())
recoTable_df.head()
```

```python
recoTable_df = recoTable_df.sort_values(ascending = False) # 내림차순
recoTable_df
```

# 결과

```python
rank = course_df.loc[course_df['courseid'].isin(recoTable_df.head(20).keys())]
rank
```

```python
rankId = data[data['course'].isin(rank['course'].tolist())]
final = pd.merge(rankId, rank)
final
```

```python
fi_result = final[['courseid', 'course', 'rating']].sort_values('rating', ascending = False) #rating순대로 배열
fi_result
```

```python
fi_result = fi_result.drop_duplicates('courseid', keep='first') #중복 제거
fi_result
```
